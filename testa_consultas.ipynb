```python
from time import time
from cassandra.cluster import Cluster
from cassandra.query import tuple_factory
import pandas as pd

cluster = Cluster(['35.203.92.156'])
cluster.control_connection_timeout = 20.0 #None
session = cluster.connect()
formato = 'utf8' 
```

```python
KEYSPACE = "bdm"
#KEYSPACE = "microdados"
session.set_keyspace(KEYSPACE)
session = cluster.connect(KEYSPACE)
session.row_factory = tuple_factory
```

```python
dfQuery = pd.read_csv('saida_light.csv', sep=";")

for index, row in dfQuery.iterrows():
    qq = row['query'] + ' LIMIT 1'
    print (str(index) + ' - ' + str(qq))
    try:
        rows = session.execute(qq, timeout=20);
        count = 0
        for i in rows:
            count += 1
            break
        print(' tem ' + str(count))

        pass
    except Exception as e: # work on python 3.x
        print("Erro - " + str(index) + ' - ' + str(e))
```
